# Pdf loader 

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import CharacterTextSplitter

# Load the document
# loader = PyPDFLoader("russian mines.pdf")
loader = PyPDFLoader("russian_Mineimages.pdf")
documents = loader.load()

# Split the document into chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30, separator="\n")
docs = text_splitter.split_documents(documents=documents)
```

```python
docs[0].page_content[:500]  # Display the first 500 characters of the first chunk
```

# Data loading and ingestion

```python
from ollama import generate
# Regular response
response = generate('llama3.1', 'Why is the sky blue?')
print(response['response'])
```

```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS

# Use a dedicated embedding model
embeddings = OllamaEmbeddings(model="nomic-embed-text")

# Create FAISS vector store
vectorstore = FAISS.from_documents(docs, embeddings)

# Save and reload the vector store
vectorstore.save_local("faiss_index_ollama")
persisted_vectorstore = FAISS.load_local("faiss_index_ollama", embeddings, allow_dangerous_deserialization=True)

# Create a retriever
retriever = persisted_vectorstore.as_retriever()
```

## Method 2: Pickle the Retriever Directly

```python
import pickle
import os
from langchain_community.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS

# File paths
retriever_path = "retriever.pkl"
vector_store_path = "faiss_index_ollama"

embeddings = OllamaEmbeddings(model="nomic-embed-text")

if os.path.exists(retriever_path):
    print("Loading existing retriever...")
    # Load existing retriever
    with open(retriever_path, 'rb') as f:
        retriever = pickle.load(f)
    print("Retriever loaded successfully!")
else:
    print("Creating new retriever...")
    # Create FAISS vector store
    vectorstore = FAISS.from_documents(docs, embeddings)
    
    # Save vector store
    vectorstore.save_local(vector_store_path)
    
    # Create retriever
    retriever = vectorstore.as_retriever()
    
    # Save retriever
    with open(retriever_path, 'wb') as f:
        pickle.dump(retriever, f)
    
    print("Retriever created and saved successfully!")
```

# Initializing LLaMA with OLLAMA

```python
from langchain_community.llms import Ollama

# Initialize the LLaMA model
llm = Ollama(model="llama3.1")

# Test with a sample prompt
response = llm.invoke("Tell me a joke")
print(response)
```

# Orchestrating RAG with RetrievalQA

```python
from langchain.chains import RetrievalQA

# Create RetrievalQA
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever)

# Interactive query loop
while True:
    query = input("Type your query (or type 'Exit' to quit): \n")
    if query.lower() == "exit":
        break
    result = qa.run(query)
    print(result)
```

# System prompt for the retriever

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate

# Define a custom prompt template for landmine identification
template = """You are a specialized explosive ordnance disposal (EOD) expert and landmine identification specialist. 
Your role is to assist defense personnel in identifying landmines based on field descriptions and available intelligence.

CRITICAL SAFETY NOTICE: Always prioritize safety protocols. If uncertain about identification, recommend contacting EOD specialists.

Use the following pieces of context to help identify the landmine type based on the field description provided. 
If you cannot make a confident identification based on the provided context, clearly state your uncertainty and recommend additional verification.
Always provide:
1. Most likely 3 landmine type and model
2. Key identifying characteristics
3. Safety considerations and disarming procedures if available
4. Confidence level in your identification

Context: {context}

Field Description/Question: {question}

Expert Analysis:"""

PROMPT = PromptTemplate(
    template=template, 
    input_variables=["context", "question"]
)

# Create RetrievalQA with custom prompt
qa = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    chain_type_kwargs={"prompt": PROMPT},
    return_source_documents=True
)

# Enhanced query loop for landmine identification
print("=== LANDMINE IDENTIFICATION SYSTEM ===")
print("Provide detailed descriptions of suspected landmines for expert analysis.")
print("Include: size, shape, color, markings, location found, etc.")
print("Type 'help' for guidance on providing descriptions.")
print("=" * 50)

while True:
    query = input("\nDescribe the suspected landmine (or type 'Exit' to quit, 'help' for guidance): \n")
    
    if query.lower() == "exit":
        break
    
    if query.lower() == "help":
        print("\n=== GUIDANCE FOR LANDMINE DESCRIPTION ===")
        print("Please provide as much detail as possible:")
        print(" Physical appearance: size, shape, color, material")
        print(" Markings: text, numbers, symbols, manufacturer marks")
        print(" Location: where found, terrain type")
        print(" Trigger mechanism: visible components")
        print(" Any other distinctive features")
        print("Example: 'Round metal disc, 10cm diameter, green color, Cyrillic markings, found in forest path'")
        continue
    
    result = qa(query)
    
    print("\n" + "="*60)
    print("EXPERT ANALYSIS:")
    print("="*60)
    print(result["result"])
    
    # Show source documents for verification
    if "source_documents" in result and result["source_documents"]:
        print("\n" + "-"*40)
        print("REFERENCE SOURCES:")
        print("-"*40)
        for i, doc in enumerate(result["source_documents"]):
            print(f"\nSource {i+1}: Page {doc.metadata.get('page', 'Unknown')}")
            print(f"Reference: {doc.page_content[:300]}...")
            if len(doc.page_content) > 300:
                print("[Content truncated...]")
    
    print("\n" + "="*60)
    print("  SAFETY REMINDER: Always follow proper EOD protocols")
    print("="*60)
```
