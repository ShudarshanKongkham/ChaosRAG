```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
import torch
import os
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Paths and model name from your code
output_dir = "./results"
model_name = 'google/gemma-2b-it'
merged_model_path = os.path.join(output_dir, "/content/results/final_adapter")
adapter_config_path = os.path.join(merged_model_path, "adapter_config.json")
adapter_model_path = os.path.join(merged_model_path, "adapter_model.safetensors")

# Function to load and chunk the text file
def load_and_chunk_text(file_path, chunk_size=1000, chunk_overlap=20):
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()
        splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
        return splitter.split_text(text)
    except Exception as e:
        print(f"Error loading text file: {e}")
        return []

# Initialize retriever and FAISS index
def initialize_retriever(text_file_path):
    documents = load_and_chunk_text(text_file_path)
    if not documents:
        raise ValueError("No documents loaded from the text file.")

    retriever_model = SentenceTransformer('all-MiniLM-L6-v2')
    document_embeddings = retriever_model.encode(documents, convert_to_numpy=True)

    dimension = document_embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(document_embeddings)

    return retriever_model, documents, index

# Initialize model and tokenizer
def initialize_model_and_tokenizer():
    if not (os.path.exists(adapter_config_path) and os.path.exists(adapter_model_path)):
        raise FileNotFoundError(f"Adapter files not found at {merged_model_path}.")

    try:
        base_model = AutoModelForCausalLM.from_pretrained(
            model_name,
            device_map="auto",
            torch_dtype=torch.float16,
            # local_files_only=True
        )
        tokenizer = AutoTokenizer.from_pretrained(
            model_name,
            # local_files_only=True
        )
        loaded_peft_model = PeftModel.from_pretrained(base_model, merged_model_path)
        loaded_peft_model.eval()
        return loaded_peft_model, tokenizer
    except Exception as e:
        raise RuntimeError(f"Error loading model or tokenizer: {e}")

# Load components once
try:
    retriever_model, documents, faiss_index = initialize_retriever("/content/combineDataset.txt")  # Replace with your text file path
    model, tokenizer = initialize_model_and_tokenizer()
    print("Model, tokenizer, and retriever initialized successfully.")
except Exception as e:
    print(f"Initialization failed: {e}")
    model, tokenizer, retriever_model, documents, faiss_index = None, None, None, [], None

```

```python
def rag_pipeline(query, top_k=4):
    if any(x is None for x in [model, tokenizer, retriever_model, faiss_index]):
        return "Error: Model, tokenizer, or retriever not initialized properly."

    # Encode query
    query_embedding = retriever_model.encode([query], convert_to_numpy=True)

    # Retrieve top-k documents
    distances, indices = faiss_index.search(query_embedding, top_k)
    retrieved_docs = [documents[idx] for idx in indices[0]]
    print(retrieved_docs)
    print("#################################################################")

    # Create prompt
    prompt = f"""<bos><start_of_turn>user
You are an expert in explosive ordnance disposal and landmine identification.
You will be given a description of a landmine’s appearance, mechanism, or markings. Your task is to identify the landmine’s name and then provide a detailed paragraph summarizing its details.
Instruction: Identify the mine and describe it based on the provided context. you can quote the context directly. give the answer in paragraph format. you can also propose 3 potential names that might be that landmine.
Question: {query}
Context: {' '.join(retrieved_docs)}<end_of_turn>
<start_of_turn>model:
"""

    # Tokenize and generate
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=1024,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )

    # Decode response
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer
```

```python
import json

# Load the provided JSON file
input_path = '/content/fine_tune_landmines.json'
with open(input_path, 'r', encoding='utf-8') as f:
    entries = json.load(f)

transformed = []
for entry in entries:
    # Parse original prompt into parts
    instr, quest, ctx = "", "", ""
    for line in entry['prompt'].splitlines():
        if line.startswith("### Instruction:"):
            instr = line.replace("### Instruction:", "").strip()
        elif line.startswith("### Question:"):
            quest = line.replace("### Question:", "").strip()
        elif line.startswith("### Context:"):
            ctx = line.replace("### Context:", "").strip()

    # Build Gemma-style prompt
    parts = [
        "<bos><start_of_turn>user Below is an instruction that describes a task. Write a response that appropriately completes the request.",
        f"{instr}",
        f"Input: {quest}"
    ]
    if ctx:
        parts.append(f"Context: {ctx}")
    parts.append("<end_of_turn>:")

    gemma_prompt = "\n\n".join(parts)

    transformed.append({
        "prompt": gemma_prompt,
        "response": "<start_of_turn>model "+entry["response"]
    })

# Save the transformed JSON
output_path = '/content/fine_tune_landmines2_gemma.json'
with open(output_path, 'w', encoding='utf-8') as f:
    json.dump(transformed, f, ensure_ascii=False, indent=2)

output_path
```
